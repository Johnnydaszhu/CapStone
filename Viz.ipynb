```python
import json
import os

from pyecharts import options as opts
from pyecharts.charts import Graph, Page
from pyecharts.faker import Collector

C = Collector()

with open(os.path.join("data.json"), "r", encoding="utf-8") as f:
    j = json.load(f)
nodes = [
    {
        "x": node["x"],
        "y": node["y"],
        "id": node["id"],
        "name": node["label"],
        "symbolSize": node["size"],
        "itemStyle": {"normal": {"color": node["color"]}},
    }
    for node in j["nodes"]
]

edges = [
    {"source": edge["sourceID"], "target": edge["targetID"]} for edge in j["edges"]
]

c = (
    Graph(init_opts=opts.InitOpts())
    .add(
        "",
        nodes=nodes,
        links=edges,
        layout="none",
        label_opts=opts.LabelOpts(is_show=False),
        linestyle_opts=opts.LineStyleOpts(width=0.5, curve=0.3, opacity=0.7),
    )
    .set_global_opts(title_opts=opts.TitleOpts(title=None))
)
c.render_notebook()
```

```python
from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.charts import Bar, Tab, Pie, Line
from pyecharts.components import Table

def bar_datazoom_slider1() -> Bar:
    c = (
        Bar()
        .add_xaxis(list(df_topic_a['word']))
        .add_yaxis("Skills Importance", list(df_topic_a['importance']))
        .reversal_axis()
        .set_series_opts(label_opts=opts.LabelOpts(position="right"))
        .set_global_opts(
            title_opts=opts.TitleOpts('Type A Analyst')
        )
    )
    return c
def bar_datazoom_slider2() -> Bar:
    c = (
        Bar()
        .add_xaxis(list(df_topic_b['word']))
        .add_yaxis("Skills Importance", list(df_topic_b['importance']))
        .reversal_axis()
        .set_series_opts(label_opts=opts.LabelOpts(position="right"))
        .set_global_opts(
            title_opts=opts.TitleOpts('Type B Analyst'),
        )
    )
    return c
def bar_datazoom_slider3() -> Bar:
    c = (
        Bar()
        .add_xaxis(list(df_topic_c['word']))
        .add_yaxis("Skills Importance", list(df_topic_c['importance']))
        .reversal_axis()
        .set_series_opts(label_opts=opts.LabelOpts(position="right"))
        .set_global_opts(
            title_opts=opts.TitleOpts('Type C Analyst'),
        )
    )
    return c
def bar_datazoom_slider4() -> Bar:
    c = (
        Bar()
        .add_xaxis(list(df_topic_d['word']))
        .add_yaxis("Skills Importance", list(df_topic_d['importance']))
#         .reversal_axis()
        .set_series_opts(label_opts=opts.LabelOpts(position="right"))
        .set_global_opts(
            title_opts=opts.TitleOpts('Type D Analyst'),
        )
    )
    return c
tab = Tab()
tab.add(bar_datazoom_slider1(), "Type A")
tab.add(bar_datazoom_slider2(), "Type B")
tab.add(bar_datazoom_slider3(), "Type C")
tab.render_notebook()
```
